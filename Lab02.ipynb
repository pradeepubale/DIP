## Digital Image Processing  

Import all the liabraries needed i.e. cv2(Computer vision 2), NumPy, Matplotlib

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

Use `imread()` function of `cv2` liabrary  
`cv2.imread()` method loads an image from the specified file. If the image cannot be read (because of the missing file, improper permissions, or unsupported or invalid format) then this method returns an empty matrix.   
Syntax: `cv2.imread(filename, flag) `  
flag - The flag specifies the way how the image should be read.  
* 1 - It specifies to load a color image  
* 0 - It specifies to load an image in grayscale mode.

```python
image = cv2.imread('0001.jpeg',1)
window_name = 'img'
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cropped_image = image[80:280,150:330]
cv2_imshow(cropped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resize_image = cv2.resize(image, [500,500])
cv2_imshow(resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resize_image = cv2.resize(image, [500,500], interpolation=cv2.INTER_NEAREST)
cv2_imshow(resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resize_image = cv2.resize(image, [500,500], interpolation=cv2.INTER_LINEAR)
cv2_imshow(resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resize_image = cv2.resize(image, [500,500], interpolation=cv2.INTER_CUBIC)
cv2_imshow(resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rotate_image = cv2.rotate(image, rotateCode=cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotate_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rotate_image = cv2.rotate(image, rotateCode=cv2.ROTATE_180)
cv2_imshow(rotate_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rotate_image = cv2.rotate(image, rotateCode=cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(rotate_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cricle_image = cv2.circle(image, [150,150],100,(255,255,255),2)
cv2_imshow(cricle_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rectangle_image = cv2.rectangle(image,[45,65],[200,150],(0,0,0),5)
cv2_imshow(rectangle_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
text_image = cv2.putText(image,'FLOWER',[50,50],font,2,(255,255,203),2)
cv2_imshow(text_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
