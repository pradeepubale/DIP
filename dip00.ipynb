```python
# First import the required Python Libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python
array_1 = np.array([[255, 0], [0, 255]], dtype=np.uint8)
imshow(array_1, cmap='gray')
```

```python
array_2 = np.array([[255, 0, 255],
                    [0, 255, 0],
                    [255, 0, 255],
                    [0, 255, 0]])
imshow(array_2, cmap='gray')
```

```python
array_spectrum = np.array([
    np.arange(0, 255, 17, dtype=np.uint8),
    np.arange(255, 0, -17, dtype=np.uint8),
    np.arange(0, 255, 17, dtype=np.uint8),
    np.arange(255, 0, -17, dtype=np.uint8)
])
fix,ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].imshow(array_spectrum, cmap='gray')
ax[0].set_title("Arange Generation")

ax[1].imshow(array_spectrum.T, cmap='gray')
ax[1].set_title("Transpose Generation")
```

```python
# array of colors

array_colors = np.array([[[255, 0, 0],    # RED
                          [0, 255, 0],    # GREEN
                          [0, 0,255]]])   # BLUE
imshow(array_colors, cmap='gray') # cmap = 'gray'
```

Let's manipulate an actual image here.

```python
kitty = imread('/content/drive/MyDrive/dip/images/cat01.jpg')
imshow(kitty)
```

```python
# to display height, width and color
print(kitty.shape)
```

```python
fig, ax = plt.subplots(1, 4,
figsize=(6,4),
sharey= True)
ax[0].imshow(kitty[:, 0:200])
ax[0].set_title('First Split')
ax[1].imshow(kitty[:, 200:400])
ax[1].set_title('Second Split')
ax[2].imshow(kitty[:, 400:600])
ax[2].set_title('Third Split')
ax[3].imshow(kitty[:, 600:800])
ax[3].set_title('Fourth Split')
```

RGB Image seperated by Color

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey =
True)
ax[0].imshow(kitty[:,:,0], cmap='Reds')
ax[0].set_title('Red')
ax[1].imshow(kitty[:,:,1],cmap='Greens')
ax[1].set_title('Green')
ax[2].imshow(kitty[:,:,2],cmap='Blues')
ax[2].set_title('Blue')
```

RGB To HSV(Hue, Saturation, Value)

```python
cat_hsv = rgb2hsv(kitty)
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(cat_hsv[:,:,0], cmap='hsv')
ax[0].set_title('Hue')
ax[1].imshow(cat_hsv[:,:,1])
ax[1].set_title('Saturation')
ax[2].imshow(cat_hsv[:,:,2])
ax[2].set_title('Value');
```

convert the image matrix to
grayscale

```python
cat_gray = rgb2gray(kitty)
fig, ax = plt.subplots(1, 5, figsize=(17,6), sharey = True)
ax[0].imshow(cat_gray, cmap = 'gray')
ax[0].set_title('GrayscaleOriginal')
ax[1].imshow(img_as_uint(cat_gray > 0.25),cmap = 'gray')
ax[1].set_title('Greater than 0.25')
ax[2].imshow(img_as_uint(cat_gray > 0.50), cmap = 'gray')
ax[2].set_title('Greater than 0.50')
ax[3].imshow(img_as_uint(cat_gray > 0.75), cmap = 'gray')
ax[3].set_title('Greater than 0.75')
ax[4].imshow(img_as_uint(cat_gray > np.mean(cat_gray)),cmap = 'gray')
ax[4].set_title('Greater than Mean')
```
